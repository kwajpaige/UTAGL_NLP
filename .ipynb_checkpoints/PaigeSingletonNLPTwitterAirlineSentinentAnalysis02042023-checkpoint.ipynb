#### Distribution of all tweets among each airline (using code from Social Dilemma Tweet Analysis)


```python
# function to create labeled barplots
def labeled_barplot(data, feature,  title, pallet,perc=True, n=None):
    """
    Barplot with percentage at the top

    data: dataframe
    feature: dataframe column
    perc: whether to display percentages instead of count (default is False)
    n: displays the top n category levels (default is None, i.e., display all levels)
    """

    total = len(data[feature])  # length of the column
    count = data[feature].nunique()
    if n is None:
        plt.figure(figsize=(16, 4))
    else:
        plt.figure(figsize=(16, 4))

    plt.xticks(rotation=90, fontsize=15)
    ax = sns.countplot(
        data[feature],
        palette=pallet,
        order=data[feature].value_counts().index[:20],
    )
    ax.set_title('Frequency of {} tweeting about Airline Customer Experience'.format(title))

    for p in ax.patches:
        if perc == True:
            label = "{:1.2f}%".format(
                100 * p.get_height() / total
            )  # percentage of each class of the category
        else:
            label = p.get_height()  # count of each level of the category

        x = p.get_x() + p.get_width() / 2  # width of the plot
        y = p.get_height()  # height of the plot

        ax.annotate(
            label,
            (x, y),
            ha="center",
            va="center",
            size=12,
            xytext=(0, 5),
            textcoords="offset points",
        )  # annotate the percentage

    plt.show()  # show the plot
```

##### Removal of numbers (using code from Trump Tweets Case Study)

```python
def remove_numbers(text):
  text = re.sub(r'\d+', '', text)
  return text

data_sm['text'] = data_sm['text'].apply(lambda x: remove_numbers(x))
data_sm.head(10)
```

###### comment: See row 5, the number 30 has been removed

##### Tokenization (using code from Trump Tweet Case Study)

```python
data_sm['text'] = data_sm.apply(lambda row: nltk.word_tokenize(row['text']), axis=1)
data_sm.head()
```

###### comment - Notice the seperation of words with commas

##### Reemoval of NON- ASCII Characters (using code from Trump Tweet Case Study)

```python
def remove_non_ascii(words):
    """Remove non-ASCII characters from list of tokenized words"""
    new_words = []                        
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)        # Append processed words to new list.
    return new_words
```

```python
data_sm['text'] = data_sm['text'].apply(lambda x: remove_non_ascii(x))
data_sm.head()
```

##### Convert all alpphabetical characters to lowercase (using code from Trump Tweet Case Study)

```python
def to_lowercase(words):
    """Convert all characters to lowercase from list of tokenized words"""
    new_words = []                        
    for word in words:
        new_word = word.lower()           
        new_words.append(new_word)        
    return new_words
```

```python
data_sm['text'] = data_sm['text'].apply(lambda x: to_lowercase(x))
data_sm.head()
```

###### Comment: Notice the decapitalization of Virgin America

##### Remove punctuation (using code from Trump Tweet Case Study)

```python
def remove_punctuation(words):
    """Remove punctuation from list of tokenized words"""
    new_words = []                        
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)    
    return new_words
```

```python
data_sm['text'] = data_sm['text'].apply(lambda x: remove_punctuation(x))
data_sm.head()
```

###### Comment: Notice the removal of punctuation marks, including the @

##### Removal of stopwords (using code from Trump Tweet Case Study)

```python

nltk.download('stopwords') 
```

```python
stopwords = stopwords.words('english')
```

```python
def remove_stopwords(words):
    """Remove stop words from list of tokenized words"""
    new_words = []                        
    for word in words:
        if word not in stopwords:
            new_words.append(word)        
    return new_words
```

```python
data_sm['text'] = data_sm['text'].apply(lambda x: remove_stopwords(x))
data_sm.head()
```

##### Lemmatization (using code from Trump Tweet Case Study)

```python
nltk.download('omw-1.4')
lemmatizer = WordNetLemmatizer()
```

```python
def lemmatize_list(words):
    new_words = []
    for word in words:
      new_words.append(lemmatizer.lemmatize(word))
    return new_words

data_sm['text'] = data_sm['text'].apply(lambda x: lemmatize_list(x))
data_sm.head()
```

###### Notice in row 1, the change from "commercials" to "commercial"

###### Comment - Notice the removal of common English words such as what, I, it, etc.....

##### Join the words in the list (using code from Trump Tweet Case Study)

```python
def normalize(words):
    words = remove_non_ascii(words)
    words = to_lowercase(words)
    words = remove_punctuation(words)
    words = remove_stopwords(words)
    words = lemmatize_list(words)
    return ' '.join(words)

data_sm['text'] = data_sm.apply(lambda row: normalize(row['text']), axis=1)
data_sm.head()
```

### Modelling , tuning and Evaluation (using code from Amazon Fine food Review Case Study)
- Fit the model using vectorized column 
- Tune the model to improve the accuracy 
- Evaluate the model using the confusion matrix 
- Target the final score >= 75% 
- Print the top 40 features and plot their word cloud using both types of vectorization. (7+7 Marks)

```python
labels = data['airline_sentiment']
```

```python
# Split data into training and testing set.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(data_features_cv, labels, test_size=0.3, random_state=42)
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import cross_val_score

forest = RandomForestClassifier(n_estimators=10, n_jobs=4)

forest = forest.fit(X_train, y_train)

print(forest)

print(np.mean(cross_val_score(forest, data_features_cv, labels, cv=10)))
```

```python
##### Comment, with 10 estimators, the model is 70% accurate on training data.
```

```python
# Finding optimal number of base learners using k-fold CV ->
base_ln = [x for x in range(1, 25)]
```

```python
cv_scores = []
for b in base_ln:
    clf = RandomForestClassifier(n_estimators = b)
    scores = cross_val_score(clf, X_train, y_train, cv = 5, scoring = 'accuracy')
    cv_scores.append(scores.mean())
```

```python
# plotting the error as k increases
error = [1 - x for x in cv_scores]                                 #error corresponds to each nu of estimator
optimal_learners = base_ln[error.index(min(error))]                #Selection of optimal nu of n_estimator corresponds to minimum error.
plt.plot(base_ln, error)                                           #Plot between each nu of estimator and misclassification error
xy = (optimal_learners, min(error))
plt.annotate('(%s, %s)' % xy, xy = xy, textcoords='data')
plt.xlabel("Number of base learners")
plt.ylabel("Misclassification Error")
plt.show()
```

```python
# Training the best model and calculating accuracy on test data .
clf = RandomForestClassifier(n_estimators = optimal_learners)
clf.fit(X_train, y_train)
clf.score(X_test, y_test)
```

```python
#####  Comment: Through optiizing the number of estimators (trees), the model increased accuracy to 75% on testing data.
```

```python
result =  clf.predict(X_test)                  #saving the prediction on test data as a result
```

```python
# Print and plot Confusion matirx to get an idea of how the distribution of the prediction is, among all the classes.

import matplotlib.pyplot as plt
import seaborn as sns
from sklearn import metrics
from sklearn.metrics import confusion_matrix

conf_mat = confusion_matrix(y_test, result)

print(conf_mat)

print(metrics.f1_score(y_test, result,average='micro'))

df_cm = pd.DataFrame(conf_mat, index = [i for i in "123"],
                  columns = [i for i in "123"])
plt.figure(figsize = (10,7))
sns.heatmap(df_cm, annot=True, fmt='g')
```
